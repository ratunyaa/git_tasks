**4-5. Рассчитайте среднее значение и среднеквадратичное отклонение возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год**

```python
# salary: >50K,<=50K
#average age by salary
avg_low_salary = (data[data.salary == '<=50K'].agg({'age':'mean'})).round(2)
print(avg_low_salary)
avg_high_salary = (data[data.salary == '>50K'].agg({'age':'mean'})).round(2)
print(avg_high_salary)
#standard deviation
sd_low_salary_age = (data[data.salary == '<=50K'].agg({'age':'std'})).round(2)
print(sd_low_salary_age)
sd_high_salary_age = (data[data.salary == '>50K'].agg({'age':'std'})).round(2)
print(sd_high_salary_age)
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
filt_m = data.query('`salary`==">50K" & `sex`=="Male" & (`marital-status` =="Married-civ-spouse" |`marital-status` =="Married-spouse-absent" |`marital-status` =="Married-AF-spouse")').count()
filt_not_m = data.query('`salary`==">50K" & `sex`=="Male" & ~(`marital-status` =="Married-civ-spouse" |`marital-status` =="Married-spouse-absent" |`marital-status` =="Married-AF-spouse")').count()
total_men = filt_m+filt_not_m
share_mar = filt_m/total_men*100
share_not_mar = filt_not_m/total_men*100
print('share of people with status "Married" is',share_mar.age.round(2),"%")
print('share of people with status "Not married" is',share_not_mar.age.round(2),"%")

#Number of people with high salary and marital status 'Married' is considerably higher 
#than people with status 'Not Married'

#variant_2
data.loc[(data['sex']=='Male')&
         (data['marital-status'].isin(['Never-married',
                                      'Separated',
                                      'Divorced',
                                      'Widowed'])), 'salary'].value_counts()

data.loc[(data['sex']=='Male')&
        (data['marital-status'].str.startswith('Married')), 'salary'].value_counts()

data['marital-status'].value_counts()
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
data.rename(columns={"hours-per-week" : "HoursPerWeek"}, inplace=True)
max_hours = data['HoursPerWeek'].max() #maximum number of hours per week
earn_min = data[data.HoursPerWeek==data['HoursPerWeek'].max()].groupby('salary'). \
agg({'salary':'count'}).salary.loc['<=50K'] #people who work max hours and earn <=50K
earn_max = data[data.HoursPerWeek==data['HoursPerWeek'].max()].groupby('salary'). \
agg({'salary':'count'}).salary.loc['>50K'] #people who work max hours and earn max salary(>50K)
print("Share of people who work max hours per week and earn a lot is", \
      (earn_max/(earn_max+earn_min)*100).round(2),'% in total number of people who work maximum number of hours')
```

**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

```python
conditions = [
             (data['age'] < 35) & (data['age'] >= 16),
             (data['age'] < 70) & (data['age'] >= 35),
             (data['age'] < 100) & (data['age'] >= 70)]
values = ['young','adult','retiree']
data['AgeGroup'] = np.select(conditions, values)
data.head(15)

# variant_2 (with cut)
cut_labels = ['young','adult','retiree']
cut_bins = [16, 35, 70, 100]
data['AgeGroup'] = pd.cut(data['age'],
                         bins = cut_bins,
                         labels = cut_labels)
data.head(15)
```
